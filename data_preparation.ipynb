9 Data Frames Creation from raw data set for Data Exploration and Analysis.

```python
# 1.aggregated_transaction data frame creation
# importing required libraries
import os
import json
import pandas as pd

aggr_trans_main_path = "dataset/data/aggregated/transaction/country/india/state/"
aggr_trans_state = os.listdir(aggr_trans_main_path)
aggr_trans_state #list of states in India

# aggregated_transaction Data Frame Structure Creation using Dictionary
aggr_trans_df_struct = { "State" : [], "Year" : [], "Quarter" : [], "Transaction_Type" : [], "Transaction_Count" : [], "Transaction_Amount" : []}

for i in aggr_trans_state:
    aggr_trans_year_path = aggr_trans_main_path + i + "/"
    aggr_trans_year = os.listdir(aggr_trans_year_path)
    aggr_trans_year
    for j in aggr_trans_year:
        aggr_trans_quarter_path = aggr_trans_year_path + j + "/"
        aggr_trans_quarter = os.listdir(aggr_trans_quarter_path)
        for k in aggr_trans_quarter:
            aggr_trans_json_path = aggr_trans_quarter_path + k
            Data = open(aggr_trans_json_path, 'r')
            D = json.load(Data)
            for z in D["data"]["transactionData"]:
                Name = z["name"]
                Count = z["paymentInstruments"][0]["count"]
                Amount = z["paymentInstruments"][0]["amount"]
                aggr_trans_df_struct["Transaction_Type"].append(Name)
                aggr_trans_df_struct["Transaction_Count"].append(Count)
                aggr_trans_df_struct["Transaction_Amount"].append(Amount)
                aggr_trans_df_struct["State"].append(i)
                aggr_trans_df_struct["Year"].append(j)
                aggr_trans_df_struct["Quarter"].append(int(k.strip(".json")))

# Data Frame Creation using Pandas Library
aggr_trans_df = pd.DataFrame(aggr_trans_df_struct)
print(aggr_trans_df)

# Storing the Data Frame in CSV format in another location for Sql scripting
aggr_trans_df.to_csv("dataframes/aggregated_transaction.csv", index=False)



    


```

```python
# 2.aggregated_user data frame creation
# aggregated_user Data Frame Structure Creation using Dictionary
aggr_usr_df_struct = { "State" : [], "Year" : [], "Quarter" : [], "User_Device" : [], "User_Count" : [],  "User_Share" : []}

aggr_usr_main_path = "dataset/data/aggregated/user/country/india/state/"
aggr_usr_state = os.listdir(aggr_usr_main_path)
aggr_usr_state

for i in aggr_usr_state:
    aggr_usr_year_path = aggr_usr_main_path + i + "/"
    aggr_usr_year = os.listdir(aggr_usr_year_path)
    for j in aggr_usr_year:
        aggr_usr_quarter_path = aggr_usr_year_path + j + "/"
        aggr_usr_quarter = os.listdir(aggr_usr_quarter_path)
        for k in aggr_usr_quarter:
            aggr_usr_json_path = aggr_usr_quarter_path + k
            usr_Data = open(aggr_usr_json_path, 'r')
            usr_D = json.load(usr_Data)
            users = usr_D["data"]["usersByDevice"]
            if users is not None: # Error Handling : Some json files may return None users
             for z in users:
                Device = z["brand"]
                Count = z["count"]
                Share = z["percentage"]
                aggr_usr_df_struct["User_Device"].append(Device)
                aggr_usr_df_struct["User_Count"].append(Count)
                aggr_usr_df_struct["User_Share"].append(Share)
                aggr_usr_df_struct["State"].append(i)
                aggr_usr_df_struct["Year"].append(j)
                aggr_usr_df_struct["Quarter"].append(int(k.strip(".json")))

# Dataframe creation using Pandas library.
aggr_usr_df = pd.DataFrame(aggr_usr_df_struct)
print(aggr_usr_df)

# Storing Data Frame in a CSV Format
aggr_usr_df.to_csv("dataframes/aggregated_user.csv", index = False)
                
```

```python
# 3. aggregate_insurance data frame creation.
# aggregate_insurance data frame structure creation.
aggr_ins_df_struct = {"State" : [], "Year" : [], "Quarter" : [], "Insurance_Count" : [], "Insurance_Amount" : []}

aggr_ins_main_path = "dataset/data/aggregated/insurance/country/india/state/"
aggr_ins_state = os.listdir(aggr_ins_main_path)
aggr_ins_state

for i in aggr_ins_state:
    aggr_ins_year_path = aggr_ins_main_path + i + "/"
    aggr_ins_year = os.listdir(aggr_ins_year_path)
    for j in aggr_ins_year:
        aggr_ins_quarter_path = aggr_ins_year_path + j + "/"
        aggr_ins_quarter = os.listdir(aggr_ins_quarter_path)
        for k in aggr_ins_quarter:
            aggr_ins_json_path = aggr_ins_quarter_path + k
            ins_Data = open(aggr_ins_json_path, 'r')
            ins_D = json.load(ins_Data)
            ins = ins_D["data"]["transactionData"]
            if ins is not None:
             for z in ins:
                Count = z["paymentInstruments"][0]["count"]
                Amount = z["paymentInstruments"][0]["amount"]
                aggr_ins_df_struct["Insurance_Count"].append(Count)
                aggr_ins_df_struct["Insurance_Amount"].append(Amount)
                aggr_ins_df_struct["State"].append(i)
                aggr_ins_df_struct["Year"].append(j)
                aggr_ins_df_struct["Quarter"].append(int(k.strip(".json")))

# Data Frame Creation using Pandas Library
aggr_ins_df = pd.DataFrame(aggr_ins_df_struct)
print(aggr_ins_df)

# Storing the Data Frame in CSV format in another location for Sql scripting
aggr_ins_df.to_csv("dataframes/aggregated_insurance.csv", index=False)
```

```python
# 4. map_transaction data frame creation.
# Data Frame Structure Creation.
map_trans_df_struct = { "State" : [], "Year" : [], "Quarter" : [], "District" : [], "Transaction_Count" : [], "Transaction_Amount" : []}

map_trans_main_path = "dataset/data/map/transaction/hover/country/india/state/"
map_trans_state = os.listdir(map_trans_main_path)
map_trans_state

for i in map_trans_state:
    map_trans_year_path = map_trans_main_path + i + "/"
    map_trans_year = os.listdir(map_trans_year_path)
    for j in map_trans_year:
        map_trans_quarter_path = map_trans_year_path + j + "/"
        map_trans_quarter = os.listdir(map_trans_quarter_path)
        for k in map_trans_quarter:
            map_trans_json_path = map_trans_quarter_path + k
            map_trans_Data = open(map_trans_json_path, 'r')
            map_trans_D = json.load(map_trans_Data)
            map_trans = map_trans_D["data"]["hoverDataList"]
            if map_trans is not None:
                for z in map_trans:
                    District = z["name"]
                    Count = z["metric"][0]["count"]
                    Amount = z["metric"][0]["amount"]
                    map_trans_df_struct["District"].append(District.replace(" district", ""))
                    map_trans_df_struct["Transaction_Count"].append(Count)
                    map_trans_df_struct["Transaction_Amount"].append(Amount)
                    map_trans_df_struct["State"].append(i)
                    map_trans_df_struct["Year"].append(j)
                    map_trans_df_struct["Quarter"].append(int(k.strip(".json")))

# Data Frame Creation using Pandas Library.
map_trans_df = pd.DataFrame(map_trans_df_struct)
print(map_trans_df_struct)

# Storing the Data Frame in CSV Format in other folder.
map_trans_df.to_csv("dataframes/map_transaction.csv", index = False)
```

```python
# 7. top_transaction data frame creation
# Data Frame structure creation
top_trans_df_struct = {
    "State": [],
    "Year": [],
    "Quarter": [],
    "District": [],
    "Transaction_Count": [],
    "Transaction_Amount": []
}

top_trans_main_path = "dataset/data/top/transaction/country/india/state/"
top_trans_state = os.listdir(top_trans_main_path)

for i in top_trans_state:
    top_trans_year_path = top_trans_main_path + i + "/"
    top_trans_year = os.listdir(top_trans_year_path)

    for j in top_trans_year:
        top_trans_quarter_path = top_trans_year_path + j + "/"
        top_trans_quarter = os.listdir(top_trans_quarter_path)

        for k in top_trans_quarter:
            top_trans_json_path = top_trans_quarter_path + k

            top_trans_Data = open(top_trans_json_path, 'r')
            top_trans_D = json.load(top_trans_Data)

            
            top_trans = top_trans_D["data"].get("districts")

            if top_trans is not None:
                for z in top_trans:
                    District = z["entityName"]
                    Count = z["metric"]["count"]
                    Amount = z["metric"]["amount"]

                    top_trans_df_struct["District"].append(
                        District.replace(" district", "")
                    )
                    top_trans_df_struct["Transaction_Count"].append(Count)
                    top_trans_df_struct["Transaction_Amount"].append(Amount)
                    top_trans_df_struct["State"].append(i)
                    top_trans_df_struct["Year"].append(j)
                    top_trans_df_struct["Quarter"].append(
                        int(k.strip(".json"))
                    )

# Data Frame Creation
top_trans_df = pd.DataFrame(top_trans_df_struct)
print(top_trans_df.head())

# Store dataframe
top_trans_df.to_csv("dataframes/top_transaction.csv", index=False)
```
